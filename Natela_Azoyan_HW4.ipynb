```python
#Task 5
incoming_dgr = dict(DG.in_degree())
max_incoming_dgr = max(incoming_dgr.values())
highest_incoming_dgr_nodes = [node for node, dgr in incoming_dgr.items() if dgr == max_incoming_dgr]
highest_incoming_dgr_nodes
```

```python
#Task 6
closeness = nx.closeness_centrality(DG)
max_closeness = max(closeness.values())
max_closeness_nodes = [node for node, centrality in closeness.items() if centrality == max_closeness]
print(f'Highest closeness has {max_closeness_nodes} node')
print(f'The value of highest closeness node is {closeness[max_closeness_nodes[0]]}')
```

Closeness centrality is a measure of how centrally located or well-connected a node is within a graph. In our case the node 2642 has the shortest average distance to all other nodes in the graph.

```python
betweenness = nx.betweenness_centrality(DG)
max_betweenness = max(betweenness.values())
max_betweenness_nodes = [node for node, centrality in betweenness.items() if centrality == max_betweenness]
print(f'Highest betweenness has {max_betweenness_nodes} node')
print(f'The value of highest betweenness node is {betweenness[max_betweenness_nodes[0]]}')
```

Betweenness centrality is a measure of the extent to which a node lies on the shortest paths between other nodes in a graph. In other words, a node with high betweenness centrality has a significant influence on the flow of information or resources between other nodes in the network. The node 1684 lies on large number of shortest paths between other pairs of nodes in the network and its score is 0.03299985344063623.

```python
eigenvector = nx.eigenvector_centrality(DG, max_iter= 700)
max_eigenvector = max(eigenvector.values())
max_eigenvector_nodes = [node for node, centrality in eigenvector.items() if centrality == max_eigenvector]
print(f'Highest betweenness has {max_eigenvector_nodes} node')
print(f'The value of highest betweenness node is {eigenvector[max_eigenvector_nodes[0]]}')
```

Betweenness centrality measures the extent to which a node lies on the shortest paths between other nodes in a graph. In our case node 2655 plays a critical role in connecting other nodes in the network. It lies on a large number of shortest paths between pairs of nodes, acting as a bridge or intermediary. 

```python
#Task 7
communities = greedy_modularity_communities(DG)
```

```python
print(f'There are {len(communities)} communities created.')
```

```python
#Task 8
largest = max(communities, key = len)
print(f'Largest community has {len(largest)} nodes.')
```

This means that there is a subgraph within a larger network where 1001 nodes are highly interconnected with each other compared to nodes outside this subgraph. 

```python
smallest = min(communities, key = len)
print(f'Smallest community has {len(smallest)} nodes.')
```

This indicates that there are relatively small but internally interconnected subgroup within the larger network that consists of 6 nodes. 

```python
#Task 9
largest = sorted(communities, key=len, reverse = True)[:3]
```

```python
subgraphs = []
for community in largest:
    subgraph = DG.subgraph(community)
    subgraphs.append(subgraph)
```

```python
for m, subgraph in enumerate(subgraphs):
    plt.figure(m + 1)
    nx.draw(subgraph, with_labels = False, node_size = 40)
    plt.show()
```

```python
DGR = DG.copy()
degree_centrality = nx.degree_centrality(DG)
top_nodes_degree = sorted(degree_centrality, key = degree_centrality.get, reverse = True)[:3]
DGR.remove_nodes_from(top_nodes_degree)
```

```python
top_nodes_betweenness = sorted(betweenness, key = betweenness.get, reverse = True)[:3]
DGR.remove_nodes_from(top_nodes_betweenness)
```

```python
top_nodes_eigenvector = sorted(eigenvector, key = eigenvector.get, reverse = True)[:3]
DGR.remove_nodes_from(top_nodes_eigenvector)
```

```python
#Task 10
influencers = []
for community in largest:
    subgraph = DG.subgraph(community)
    centrality = nx.degree_centrality(subgraph)
    influencer = max(centrality, key = centrality.get)
    influencers.append(influencer)
```

```python
influencer_subgraphs = []
for influencer in influencers:
    influencer_subgraph = DG.subgraph(nx.ego_graph(DG, influencer, radius = 1))
    influencer_subgraphs.append(influencer_subgraph)
```

```python
node_sizes = []
for subgraph in influencer_subgraphs:
    centrality = nx.degree_centrality(subgraph)
    sizes = [80 if node == max(centrality, key = centrality.get) else 30 for node in subgraph.nodes()]
    node_sizes.append(sizes)
```

```python
for i, subgraph in enumerate(influencer_subgraphs):
    plt.figure(i+1)
    node_colors = ['lightblue' if node != influencers[i] else 'red' for node in subgraph.nodes()]
    nx.draw(subgraph, with_labels=False, node_color=node_colors, node_size=node_sizes[i])
    plt.title(f"Top Community {i+1}")
    plt.show()
```
